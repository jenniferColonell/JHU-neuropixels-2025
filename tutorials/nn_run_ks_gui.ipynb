## Running the kilosort gui

To sort the data, we need:
- the preprocessed binary file (in particular tshift and artifact removal are not performed by Kilosort)
  - Jan_2026_example_data\SC035_preprocessed\SC035_010720_ex_g0\SC035_010720_ex_g0_tcat.imec0.ap.bin
- the channel map, specifying which sites are recorded by each channel
  - Jan_2026_example_data\SC035_preprocessed\SC035_010720_ex_g0SC035_010720_ex_g0_tcat.imec0.ap_chanMap.mat

A spike sorting pipeline will usually generate the channel map automatically. For manual sorting of SpikeGLX data, you can use [SGLXMetaToCoords](https://github.com/jenniferColonell/SGLXMetaToCoords) or Spikeinterface.

Open a new Miniforge terminal window and activate the environment: 

``conda activate npix-analysis``

Kilosort has been installed in the environment with all of its dependencies. The installation instructions are [here.](https://github.com/MouseLand/Kilosort)

To sort with all default parameters, only the binary file, an output folder, and channel map need to be provides. Start Kilosort with:

``python -m kilosort``

![KS_gui](ks_gui.png)


Specify the binary, destination, and probe layout. If the "LOAD" button doesn't become active, try closing the gui and trying again. Once the data is able to load, you can see it before/after whitening.

Click 'Run' to start sorting.
